```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Source code from the source file **Scores_dict.py**

```python
scores = {"Day 1": 100, "Day 2": 108, "Day 3":112, "Day 4":115, "Day 5":150,
          "Day 6":178, "Day 7": 143, "Day 8": 132, "Day 9":190, "Day 10": 235,
          "Day 11":253, "Day 12": 298, "Day 13": 328, "Day 14":390, "Day 15": 257,
          "Day 16":288, "Day 17": 393, "Day 18": 425, "Day 19":458, "Day 20": 450,
          "Day 21":473, "Day 22": 333, "Day 23": 452, "Day 24":490, "Day 25": 495,
          "Day 26":488, "Day 27": 543, "Day 28": 532, "Day 29":590, "Day 30": 605}
```

> The solution is obtained step by step as instructed <br> in the given problem.

1. Array named *Scores[ ]* extracted from the given dictionary **scores**

```python
Scores = np.array(list(scores.values())).reshape(-1, 1)
```

```python
Scores
```

2. With the help of numpy function **arange()** another array *Days[ ]* containing numbers <br> from 1 to 30 is obtained.

```python
Days = np.arange(1, 31).reshape(-1, 1)
```

```python
Days
```

3. A **Score vs Days** graph is obtained with X-axis labelled as *Score* and Y-axis labelled as *Days*.

```python
fig, ax = plt.subplots(figsize = (12, 8))
ax.plot(Scores, Days, 'o-r', label = 'Score vs Days', linewidth = 2)
ax.legend(loc = 'best', fontsize = 'xx-large', shadow = True)
ax.set_xlabel('Score', fontsize = 'xx-large')
ax.set_ylabel('Days', fontsize = 'xx-large')
ax.tick_params(length = 10, labelsize = 'large')
ax.grid(linewidth = 2, alpha = 0.5)
fig.suptitle('Score vs Days', fontsize = 30, fontweight = 'bold')
plt.show()
```

>The above graph signifies overall **increase** in the performance of Billy.
>
>There are **three** major setbacks in Billy's performance where his score reduced significantly from the previous scores.

4. **Mean, Median, Min, Max** of Billy's Scores are obtained with the help of *numpy function*.

```python
print('Mean Score of Billy is {:.2f}.'.format(np.mean(Scores)))
print('Median of Scores of Billy is {:.2f}.'.format(np.median(Scores)))
print('Billy has a Minimum score of {} and a Maximum score of {}.'.format(np.min(Scores), np.max(Scores)))
```

> Billy's **mean score** 333.47 is greater than his **median score** 330.50 <br> which indicates that Billy's average performance is satisfactory.
